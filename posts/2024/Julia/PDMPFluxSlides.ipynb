### How to Use? (1/3)

```julia
using PDMPFlux

function U_Gauss(x::Vector)
    return sum(x.^2) / 2
end

dim = 10
sampler = ZigZagAD(dim, U_Gauss)  # Instantiate a sampler
```

**Inputs**: Dimension $d$, and *any* $U$ that satisfies

$$
p(x) \propt \exp\left\{ -U(x) \right\},\qquad x\in\R^d.
$$

### How to Use? (2/3)


````julia
N_sk, N, xinit, vinit = 1_000_000, 1_000_000, zeros(dim), ones(dim)  # Hyperparameters
samples = sample(sampler, N_sk, N, xinit, vinit, seed=2024)
````

**Output**: A list $\{x_i\}_{i=1}^{\texttt{N_{sk}}}$ of [orange points]{.color-unite} below:
